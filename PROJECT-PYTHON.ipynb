### preprocess the data


```python
df.columns
#removing leading spaces in the column names
df.columns=df.columns.str.strip()
```

```python
#replace incorrect values in the'height' column
df["Height"]=np.random.randint(150,181, size=df.shape[0])
df.to_csv('df.csv',index=False)
df
```

```python
#remove rows with missing values in essential columns
essential_columns=['Name','Team','Position','Age','Salary']
                               
df=df.dropna(subset=essential_columns)   
df     
                  
```

### exploratory data analyis


```python
# Generate descriptive statistics
numeric_columns = ["Age", "Height", "Weight", "Salary"]
numeric_data = df[numeric_columns]
numeric_data.describe()
```

```python
# Visualize distributions using histograms
numeric_data.hist(bins=10, figsize=(12, 8))
plt.tight_layout()
plt.show()
```

```python


# Heatmap of correlations between numerical variables
corr_matrix = df[numeric_columns].corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

```

```python
# Generate value counts for categorical variables
categorical_columns = ["Team", "Position", "College"]
for column in categorical_columns:
    value_counts = df[column].value_counts()
    print(f"Value counts for {column}:")
    print(value_counts)
    print()
```



### How many employees are there in each team and the percentage splitting with respect to the total employees

```python
team_counts=df["Team"].value_counts()


print('Team_counts:') 
print(team_counts)
      
```

```python
team_percentage=(team_counts/df.shape[0])*100
print('Team percentages')
print(team_percentage)
```

### seggregate the employees w.r.t different positions

```python
position_counts=df['Position'].value_counts()
print('position_counts')
     
print(position_counts)
        
```

### determine the age group with heighest number of employees

```python
#categorize
age_groups=pd.cut(df['Age'],bins=[18,25,30,40,50,60],labels=['18-25','25-30','30-40','40-50','50-60'])
age_groups.value_counts()
```

```python
age_group_counts=age_groups.value_counts()
most_common_age_group=age_group_counts.idxmax()
print("age group with most common employee:")
print( most_common_age_group)
```

### find the team and position with the heighest salary spending

```python
heighest_salary=df.groupby(['Team','Position'])['Salary'].sum().idxmax()
heighest_salary_team,heighest_salary_position=heighest_salary
print("Team and position with heighest salary")
print("Team:",heighest_salary_team)
print("Position:",heighest_salary_position)
print()
```

### correlation between age and salary

```python
# Correlation between age and salary
df['Age'] = df['Age'].astype(int)  # Convert age column to integer
df.plot.scatter(x='Age', y='Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Correlation between Age and Salary')
plt.show()
```
